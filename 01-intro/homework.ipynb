## Q6. Median value of horsepower

```python
df['horsepower'].median()
```

```python
tmp = df['horsepower'].value_counts().items()
value, counts = next(tmp)
print(f"Most common value is: {value} with nr of counts {counts}")
```

```python
a = df['horsepower'].mode()[0]
a
```

```python
new_median = df['horsepower'].fillna(a).median()
```

```python
print(f"Median before filling in nan: {df['horsepower'].median()}")
print(f"Median after filling in nan: {new_median}")
```

```python
tmp_df = df[(df['origin'].str.lower()=="asia")].loc[:,['vehicle_weight','model_year']].iloc[:7]
tmp_df
```

```python
X = tmp_df.values
X.shape
```

```python
XTX = X.T @ X
print(XTX.shape)
print(XTX)
```

```python
XTX_inv = np.linalg.inv(XTX)
print(XTX_inv.shape)
print(XTX_inv)
```

```python
y = np.array([1100,1300,800,900,1000,1100,1200])
y.shape
```

```python
w = (XTX_inv @ X.T @ y)
w.shape
```
